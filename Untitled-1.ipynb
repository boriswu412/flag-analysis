```python
from qiskit import QuantumCircuit
from qiskit.visualization import circuit_drawer

# Load the QASM file using Qiskit
qasm_file = "[[7,1,3]]_flag.qasm"
qc = QuantumCircuit.from_qasm_file(qasm_file)

# Draw the circuit and save it as an image
circuit_drawer(qc, output="mpl", filename="circuit_diagram.png")
print("Circuit diagram saved as 'circuit_diagram.png'")
```

```python
from flag_analysis import *
qasm = "[[7,1,3]]_flag.qasm"

# sets of ancilla and flag qubits
anc_idxs  = [0,1,2,3,4,5,6]  # list of ancillas
flag_idxs = [0,1,2,3,4,5,6]   # list of flags

# build the Pauli-flow state with no faults
state, qc = build_state_from_qasm(qasm)

# extract syndrome bits (ancilla Z values) and flag bits (flag X-flip formulas)
syn = ancillas_Z(state, anc_idxs)
flg = flags_X(state, flag_idxs)

print("Syndrome results (Z on ancillas):", syn)
print("Flag results (X-basis flip formulas):", flg)
```

```python
state, qc = build_variable_state_from_qasm("[[7,1,3]]_flag.qasm")
groups = detect_qubit_groups(qc)

print("Ancillas X basis:", ancillas_X(state, groups["ancX"]))
print("Ancillas Z basis:", ancillas_Z(state, groups["ancZ"]))
print("Flags X basis:", flags_X(state, groups["flagX"]))
print("Flags Z basis:", flags_Z(state, groups["flagZ"]))
print("Data qubits:", data_qubits(state, groups["data"]))
```
